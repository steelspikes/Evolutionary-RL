```python
from functions import reconstruct_weights, params_shape, IndividualNN
import numpy as np
import gymnasium as gym
```

```python
best_NN = np.load("best_params.npy")
w, b = reconstruct_weights(params_shape, best_NN)
ind = IndividualNN(w, b)
```

```python
env = gym.make("LunarLander-v3")

rewards = []

best_reward = -np.inf

for episode in range(1000):
    observation, info = env.reset()
    terminated = False
    truncated = False
    total_reward = 0
    state = observation

    while not (terminated or truncated):
        ind_res = ind.forward(state)
        action = np.argmax(ind_res)

        next_observation, reward, terminated, truncated, info = env.step(action)

        state = next_observation
        
        total_reward += reward

    rewards.append(total_reward)

print(f"Epoch {episode + 1} | Best reward: {max(rewards)} | Mean reward: {np.mean(rewards)} | Min reward: {min(rewards)}")

env.close()
```
