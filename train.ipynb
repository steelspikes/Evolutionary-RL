```python
import gymnasium as gym
import numpy as np
import cma
from functions import calc_number_parameters, evaluate_individuals, init_worker
import multiprocessing
```

```python
if __name__ == "__main__":
    n_params = calc_number_parameters()

    opts = {
        'popsize': 32,
        'maxiter': 400,
        # 'tolfun': 1e-3,
        # 'tolx': 1e-3,
    }

    es = cma.CMAEvolutionStrategy(n_params * [0], 0.4, opts)

    while not es.stop():
        solutions = es.ask()
        with multiprocessing.Pool(processes=8, initializer=init_worker) as pool:
            costs = evaluate_individuals(pool, solutions)
    
        es.tell(solutions, costs)

        if es.countiter % 10 == 0:
            mejor_valor = es.best.f * -1
            print(f"Gen {es.countiter} | Best: {mejor_valor:.4f} | Sigma: {es.sigma:.4f}")
```

```python
# Save best individual parameters
np.save("best_params.npy", es.best.x)
```
